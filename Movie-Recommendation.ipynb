```python
import ast
```

<div class="alert alert-block alert-success">
<b>Fuction for extract genres names  : </b> so we make a function that extarct only names from the rows 
</div>

```python
def extract_genres_names(obj):
    genres_names=[]
    for i in ast.literal_eval(obj):
        genres_names.append(i['name'])
    return genres_names
```

```python
new_movies_data['genres']=new_movies_data['genres'].apply(extract_genres_names)
```

<div class="alert alert-block alert-success">
<b>Fuction for extract keywords names  : </b> so we also make a function that extarct only names from the rows for keywords.
</div>

```python
def extract_keyword_names(obj):
    keyword_names=[]
    for i in ast.literal_eval(obj):
        keyword_names.append(i['name'])
    return keyword_names
```

```python
new_movies_data['keywords']=new_movies_data['keywords'].apply(extract_keyword_names)
```

<div class="alert alert-block alert-info">
<b>Tip:</b> overview  column is a string  so we need to covert into a list of strings that we can apply sum manipulation on them 
</div>

```python
new_movies_data['overview']=new_movies_data['overview'].apply(lambda x:x.split())
```

<div class="alert alert-block alert-danger">
<b>Headache:</b> there are many garbages in cast and crew  columns so we need  some cast names  and crew names  .i personally prefer first 5 cast name and director name  from crew .you may choose how many you want its on you .   
</div>

<div class="alert alert-block alert-success">
<b>Fuction for extract top 5 cast names  : </b> so we  make a function that extarct only top 5 cast  names from the rows 
</div>

```python
def top_5_cast(obj):
    cast=[]
    for i in ast.literal_eval(obj):
        if len(cast)==5:
            break
        else:
            cast.append(i['name'])
    return cast
```

```python
new_movies_data['cast']=new_movies_data['cast'].apply(top_5_cast)
```

<div class="alert alert-block alert-success">
<b>Fuction for extract director name  : </b> so we also make a function that extarct only director name  from crew.
</div>

```python
def extract_director(obj):
    director=[]
    for i in ast.literal_eval(obj):
        if i['job']=="Director":
            director.append(i['name'])
            break
    return director   
```

```python
new_movies_data['crew']=new_movies_data['crew'].apply(extract_director)
```

### ðŸ”  Text Normalization: Porter Stemming

<div style="background-color:#1e1e1e; padding:15px; border-radius:10px; color:#f8f8f2; font-family:monospace;">
<pre>
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
</pre>
</div>

<div class="alert alert-block alert-warning">
<b>Why Stemming?</b> 
In our tags, we might have words like <i>"loving"</i>, <i>"loved"</i>, and <i>"love"</i>. 
<b>PorterStemmer</b> chops off the suffixes and brings words down to their root form (e.g., 'dancing' becomes 'danc'). This helps our system group similar words together.
</div>

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join([ps.stem(x)for x in x.split()]))
```

```python
new_df.head(1)
```

### ðŸ”¢ Count Vectorization (Bag of Words)

<div class="alert alert-block alert-success">
<b>Vectorization:</b> Since machines don't understand text, we convert our 5000 most common tags into numbers (vectors). This allows us to calculate distances between movies.
</div>

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

<div class="alert alert-block alert-success">
<b>success:</b> train the model to get that 5000 most frequent words and make a vocabulary.
</div>

```python
cv.fit(new_df['tags'])
```

<div class="alert alert-block alert-success">
<b>Vocabulary:</b> 5000 words
</div> 

```python
cv.vocabulary_
```

<div class="alert alert-block alert-success">
<b>Vocabulary to Vector:</b>  convert vocabulary to vector matrix  with respect to every movie
</div>

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

### ðŸ“ Calculating Cosine Similarity

<div class="alert alert-block alert-info">
<b>Cosine Similarity:</b> We calculate the angle between movie vectors. The smaller the angle, the more similar the movies are considered to be!
</div>

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
similarity
```

### ðŸš€ Final Recommendation System

<div class="alert alert-block alert-success">
<b>Final Result:</b> Our engine is ready! This function finds the 5 closest movies to your favorite one based on the distance calculated earlier.
</div>

```python
def movies_recon_list(movie):
    movie_name=[]
    idx=new_df[new_df['title']==movie].index[0]
    movies_list=sorted(list(enumerate(similarity[idx])),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        movie_name.append(new_df['title'][i[0]])
    return movie_name
```



<div class="alert alert-block alert-success">
<b>Final Result samples :</b> Our engine is ready! This function finds the 5 closest movies of Avatar,Interstellar,spider-man based on the distance calculated earlier.
</div>

```python
movies_recon_list("Avatar")
```

```python
movies_recon_list('Interstellar')
```

```python
movies_recon_list('Spider-Man')
```
